# Seaborn scatter plot
## Trying to get the bubble sizes to be based on driver_count. Everything else works correctly.

```python
from matplotlib import pyplot as plt
from scipy import stats

import numpy as np
import pandas as pd
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
#read the files
city_summary = pd.read_csv("small_test_seaborn.csv")
```

```python
city_summary
```

```python
# 1) Colors work with hue specifying them by city type, but bubble sizes do NOT correspond to the driver_count
      
colors = ["lightcoral", "#c6fcff", "#dbb40c"]

# these are too small to see, so used a multiplier: markersize =  city_summary['driver_count']
markersize = 30* city_summary['driver_count']

kws = {"edgecolor":'black','linewidths':1,"s": markersize}

#hue specifies to use different colors based on the type
ax = sns.lmplot(x="number_of_rides", y="avg_fare", data=city_summary, hue="type",fit_reg=False,palette=colors,
        legend_out=False,size=8,aspect=1.5, hue_order=["Urban","Suburban","Rural"],
               scatter_kws=kws) 

plt.legend(title = "City Types")
plt.title("Pyber Ridesharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare (Dollars)")

plt.show()

```

```python
# these are too small to see, so used a multiplier: markersize =  city_summary['driver_count']
markersize = 30* city_summary['driver_count']
kws = {"edgecolor":'black','linewidths':1,"s": markersize}

#Same scatter plot, same data, except do not use hue to set the colors by type
ax = sns.lmplot(x="number_of_rides", y="avg_fare", data=city_summary, fit_reg=False,
        size=8,aspect=1.5, scatter_kws=kws) 

plt.title("Pyber Ridesharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare (Dollars)")

plt.show()
```
